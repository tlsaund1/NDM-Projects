## Question 1

```python
#using BeautifulSoup
import requests
from bs4 import BeautifulSoup
```

```python
#create url object
url = 'https://books.toscrape.com'
response = requests.get(url)
soup = BeautifulSoup(response.content,'html.parser')
print(soup)
```

```python
#Find the title of the webpage
#.title to extract info inside title tag
#.text to get text content AFTER tag
#.strip() removes whitespace before and after text
soup.title.text.strip()
```

## Question 2

```python
#use soup.select to extract the tags that have attribute side_categories'
#and inside that content the ul tag, then child li tag, then child a tag
#for each result, print the text inside the tags and strip any whitespace
print(f'The categories of books from the page are:')
for i in soup.select('.side_categories ul li a'):
    print(i.text.strip())
```

## Question 3

```python
#'.side_categories ul li a' could be replaced by '.nav-list li a' to achieve same result
for i in soup.select('.side_categories ul li a'):
    print( f'Category: {i.text.strip()}, Link: {i.get("href")}') #print categories and corresponding link
```

```python
#ONLY retrieve romance category link
for i in soup.select('.side_categories ul li a'):
    if i.text.strip() == "Romance":
        print(f'romance_url is {i.get("href")}')
```
