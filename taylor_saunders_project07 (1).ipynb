Write a function that takes a year (not necessarily 2020) and a station_id as inputs, and the function creates a box plot about the month-by-month wind speeds in that specific year (not necessarily 2020), at the specified station_id.

```python
def get_wind(station_id, myyear):
    """
    returns: creates a box plot about the month-by-month wind speeds in that specific year at the specified station_id.
    """
    myresults = myDF[(myDF['station_id'] == my_station_id) & (myDF['year'] == myyear)]
    wind_box = myresults.boxplot(column = 'wind_speed_mph', by = 'month')
    return wind_box
```
